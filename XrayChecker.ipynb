```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
%config InlineBackend.figure_format = 'svg'
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
model = load_model('XceptionModel.hdf5')
```

```python
dict = {
    0 : "Normal",
    1 : "Doubtful",
    2 : "Mild",
    3 : "Moderate",
    4 : "Severe",
}
```

```python
import cv2
def SeverityChecker(img):
    resize = tf.image.resize(img, (160, 335))
    plt.imshow(resize.numpy().astype(int))
    plt.show()
    yhat = model.predict(np.expand_dims(resize / 255, 0))
    yhat = np.argmax(yhat, axis = 1)
    if yhat[0] in dict:
        print(dict[yhat[0]])
    print(yhat)
```

```python
img = cv2.imread('Testing/4Severe/4.png')
SeverityChecker(img)
```

```python
img = cv2.imread('Testing/1Doubtful/4.png')
SeverityChecker(img)
```

```python
img = cv2.imread('Testing/2Mild/4.png')
SeverityChecker(img)
```

```python
img = cv2.imread('Testing/3Moderate/4.png')
SeverityChecker(img)
```
